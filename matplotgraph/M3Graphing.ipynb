```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dfcon = pd.read_csv('ConstructionTimeSeriesDatav2.csv')

fig, ax = plt.subplots(figsize=(10, 10))
#fig, ax = plt.subplots()

ax.plot(dfcon['Month-Year'],dfcon['Private Construction'], c='k')
ax.plot(dfcon['Month-Year'],dfcon['Public Construction'], c='b')

ax.set_xticks(dfcon['Month-Year'][::15])
ax.set_xticklabels(dfcon['Month-Year'][::15])

fig.suptitle("Private and Public Construction Spending Over Time",fontsize='18',fontname='Times New Roman')
ax.set_xlabel('Year and Month',fontsize='14',fontname='Times New Roman')
ax.set_ylabel('Spending Over Time',fontsize='14',fontname='Times New Roman')

ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')
ax.set_ylim(0,max(dfcon['Private Construction'])+5000)

ax.axhline(max(dfcon['Private Construction']),linewidth=1,color='k',label='PrivHigh',linestyle='--')
#ax.axhline(min(dfcon['Private Construction']),linewidth=1,color='k',label='PrivLow',linestyle='--')
#ax.axhline(max(dfcon['Public Construction']),linewidth=1,color='b',label='PubHigh',linestyle='--')
ax.axhline(min(dfcon['Public Construction']),linewidth=1,color='b',label='PubHigh',linestyle='--')
ax.legend(loc=1)

#ax.spines['right'].set_visible(False)
#ax.spines['top'].set_visible(False)

ax.axvspan(95,120,facecolor='k',alpha=0.1)
ax.text(108,70000,'Nov-09\nthrough Nov-11',fontsize=14,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')

plt.show()
```

Data insights I am trying to provide:

The data that I wanted to highlight in this visual cover two main points: The difference in the highs and lows of public construction spending and private, and the large decline in private construction spending in 2009. I added horizontal lines that would highlight both the min and max values for both facets of construction spending; I only included the high for private and the low for public to show the disparity. The mins and maxes for each line can be provided but they cluttered the graph. I believe this is a good prepresentation of showing how different the min and max are as visually, anyone in the audience would be able to tell the distance between the lines. For the two years from November 2009 and November 2011, there is a noticeable dropoff in private construction spending. I highlighted these years by using a gray background and including text to let the audience know what years are specifically being looked at.

Graph notes:

For the graph itself, I graphed the private construction spending in black and the public construction spending in blue. I set the tick marks for the x axis to gap every 15 units, which would be 15 months, to declutter the bottom. This can be adjusted to show more by decreasing the value or increasing the size of the graph itself. After this I appropriately named the x and y axis as well as the graph overall. I changed the tick marks to follow the units on both axes and face inward towards the graph. For the horizontal lines, the other lines can be included by removing the comment # in front. I added a legend to describe what these horizontal values are.
